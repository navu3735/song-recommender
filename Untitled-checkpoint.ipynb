```python
#Import libraries 
import pandas as pd 
import numpy as np 
```

```python
#Read csv files and display
songs = pd.read_csv('Music.csv')
songs.head(1)
```

```python
#Important metrics for recommendation 
#Danceability 
#Energy 
#Valence 
#Acousticness
#Instrumentalness
#Loudness 
songs = songs[['name', 'artist', 'spotify_id', 'danceability', 'energy', 'valence', 'acousticness', 'instrumentalness', 'loudness']]
songs.head(1)
```

```python
#Drop songs with null values
songs.dropna(inplace=True)
#Drop duplicate songs 
songs = songs.drop_duplicates()
```

```python
#Get subset of songs
songs_subset = songs.sample(n=7500, random_state=0) 
songs_subset = songs_subset.reset_index(drop=True)
```

```python
#Import cosine similarity 
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Compute similarities
features = ['danceability', 'energy', 'valence', 'acousticness', 'instrumentalness', 'loudness']
feature_matrix = songs_subset[features].values
similarity_matrix = cosine_similarity(feature_matrix)
```

```python
# Function to recommend similar songs
def recommend(song_name, num_recommendations=5):
    try:
        # Find song index
        song_idx = songs_subset[songs_subset['name'] == song_name].index[0]
        
        # Get similarity scores
        sim_scores = similarity_matrix[song_idx]
        
        # Get indices of most similar songs
        similar_indices = sim_scores.argsort()[::-1][1:num_recommendations+1]  # Exclude self
        
        # Return recommendations
        return songs_subset.iloc[similar_indices][['name', 'artist']]
    except IndexError:
        return "Song not found in the 7500-song subset"
```

```python
print(recommend('La Cartera'))
```
